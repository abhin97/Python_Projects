```python
# YOUR SECTION B ANSWER HERE (please use comments to number each of your answers)

#Q1:Rename column 'long' as 'longitude', and 'lat' as 'latitude'
df.rename(columns={'long':'longitude','lat':'latitude'}, inplace=True)

##Q2:Print the number of rows for the subset of the dataframe for which traffic values are higher than
#    15,000 and lower than 25,000. Please print those rows as well.

traf_subset=df[(df["cnt"] > 15000) & (df["cnt"] < 25000)]
print(traf_subset.shape[0]) 
print(traf_subset)
print(df.columns)

##Q3:Print the names of all the columns of the dataframe
print(df.columns)

##Q4:Print a subset of the dataframe containing 
#   the 100th and the 10th rows of the original dataframe (in that order)

one_100=df.iloc[[99,9]] #df first row starts with index 0
print(one_100)

##Q5:Which US state has the largest total airport traffic? 
#   Which state has the least total airport traffic?
df['state_totals'] = df.groupby('state')['cnt'].transform('sum')
max= df.loc[df['state_totals'].idxmax()]['state']
print(max)
min= df.loc[df['state_totals'].idxmin()]['state']
print(min)

##Q6:Which state has the most total number of airports?
df['airport_count'] = df.groupby('state')['airport'].transform('nunique')
max_arpt_state= df.loc[df['airport_count'].idxmax()]['state']
print(max_arpt_state)

##Q7: Of the states that have at least 9 airports, 
#    which one has the least total traffic for all airports?
n9_arpt=df[(df["airport_count"] >= 9)]
print(n9_arpt.loc[n9_arpt['state_totals'].idxmin()]['state'])

##Q8: What is the total airport traffic for all states that border the state of Virginia?
Boader_of_virginia=np.array(["NC","TN","KY","WV","MD"])
Boader_of_virginia=df[df['state'].isin(Boader_of_virginia)]
print(Boader_of_virginia['cnt'].sum(axis=0))

##Q9: Generate a new column 'intl'- a binary feature which stores 'True' or 'False' based on whether
#     it's an international airport or not. And find which state has the most number of international airports.
df['intl'] = np.where(df['airport'].str.contains("Intl|International", case=False), True, False)
df['intl_airport_count']= df.groupby('state')['intl'].transform('sum')
max_state_intlairport= df.loc[df['intl_airport_count'].idxmax()]['state']
print(max_state_intlairport)
```
